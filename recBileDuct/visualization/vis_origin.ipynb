```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 将数据转化为DataFrame格式
data = pd.DataFrame({
    'Metric': ['Dice', 'Jaccard', 'Precision', 'Recall', 'HD95', 'ASSD'] * 2,
    'Value': [75.54, 60.87, 71.08, 81.09, 8.43, 2.27, 77.00, 62.71, 73.09, 81.55, 5.05, 1.46],
    'Network': ['Network 1'] * 6 + ['Network 2'] * 6
})

# 使用seaborn库画小提琴图
sns.set(style='whitegrid', palette='pastel')
sns.violinplot(x='Metric', y='Value', hue='Network', data=data, split=True, inner='quart', linewidth=1, 
               scale='area', dodge=True)

# 添加标题和标签
plt.title('Comparison of Segmentation Metrics between Network 1 and Network 2')
plt.xlabel('Metrics')
plt.ylabel('Values')

# 调整图像大小和显示
plt.gcf().set_size_inches(10, 6)
plt.show()
```
